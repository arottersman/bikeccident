```python
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.

    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))
```

```python
def bar_chart(x_data, y_data, x_label, y_label, title):
    sns.set(style='white')
sns.palplot(sns.color_palette("hls", 8))
    axes = sns.barplot(x_data, y_data)
    autolabel(axes.patches, axes)

    # Finalize the plot
    sns.axlabel(x_label,y_label)
    sns.despine(bottom=True)
    plt.setp(axes, xticks=[])
    plt.tight_layout(h_pad=3)
```

```python
import operator
def graph_contributing_factors(collisions):
    contributingFactors = dict()
    for collision in collisions:
        cf = "Not Reported" if collision.contributingFactor is '' else collision.contributingFactor 
        if(cf in contributingFactors):
            contributingFactors[cf] += 1
        else:
            contributingFactors[cf] = 1
    keys = list()
    values = list()
    for key, value in sorted(contributingFactors.iteritems(), key=operator.itemgetter(1)):
        keys.append(key)
        values.append(value)
    print len(keys)
    print len(values)
    print keys[len(keys)-1]
    bar_chart(values, keys,"Collisions", "Contributing Factors", "Contributing Factors")
    
```

```python
graph_contributing_factors(collisions)
```

```python
def graph_hotspot_streets(collisions, cutoff = 15):
    streets = dict()
    for collision in collisions:
        if(collision.location.crossStreetOne in streets):
            streets[collision.location.crossStreetOne] += 1
        else:
            streets[collision.location.crossStreetOne] = 1
        if(collision.location.crossStreetTwo in streets):
            streets[collision.location.crossStreetTwo] += 1
        else:
            streets[collision.location.crossStreetTwo] = 1
    keys = list()
    values = list()
    count = 0
    for key, value in reversed(sorted(streets.iteritems(), key=operator.itemgetter(1))):
        keys.append(key)
        values.append(value)
        count +=1
        if(count>cutoff):
            break
    bar_chart(values, keys,
                "Collisions", "Streets", "Contributing Factors")
```

```python
graph_hotspot_streets(collisions)
```

```python
def graph_vehicle_type(collisions, cutoff = 15):
    vehicles = dict()
    for collision in collisions:
        addedBike = False
        for vehicle, vehicleSummary in collision.involvedVehicles:
            if(vehicleSummary in vehicles):
                if(vehicleSummary != "Bicycle" 
                   or (vehicleSummary == "Bicycle" and addedBike == True)):
                    vehicles[vehicleSummary] += 1
            else:
                vehicles[vehicleSummary] = 1
                if(vehicleSummary == "Bicycle"):
                    addedBike = True
    keys = list()
    values = list()
    count = 0
    for key, value in reversed(sorted(vehicles.iteritems(), key=operator.itemgetter(1))):
        keys.append(key)
        values.append(value)
        count +=1
        if(count>cutoff):
            break
    bar_chart(values, keys,
        "Collisions", "Vehicles", "Contributing Factors")
```

```python
graph_vehicle_type(collisions)
```
